 - 서울 기온 데이터 히스토그램

```python
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)

temp = []

for row in data :
    if row[-1] != '':
        temp.append(float(row[-1]))

plt.hist(temp,color='red',bins=100)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv','r',encoding='cp949')
data = csv.reader(f,delimiter=',')
header = next(data)

temp=[]

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            temp.append(float(row[-1]))

plt.title('08 temp')
plt.hist(temp,color='red',bins=100)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv','r',encoding ='cp949')
data = csv.reader(f,delimiter=',')
header = next(data)

aug=[] #8월
jan=[] #1월

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month =='08':
            aug.append(float(row[-1]))
        elif month == '01':
            jan.append(float(row[-1]))
            
plt.title('08 temp')
plt.hist(aug,color='red',bins=100)
plt.show()

plt.title('01 tmep')
plt.hist(jan, color='blue', bins=100)
plt.show()

plt.hist(aug,color='red', label='Aug',bins=100)
plt.hist(jan,color='blue', label='Jan',bins=100)
plt.legend()
plt.show()
```

```python
import csv

f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)

aug = [] #8월
jan = [] #1월

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08' :
            aug.append(float(row[-1]))
        elif month =='01':
            jan.append(float(row[-1]))

import matplotlib.pyplot as plt
            
plt.boxplot(aug)
plt.boxplot(jan)
plt.show()

plt.boxplot([aug,jan])
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f,delimiter =',')
header = next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data :
    if row[-1] !='':
        month[int(row[0].split('-')[1])-1 ].append(float(row[-1]))
        
plt.boxplot(month)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f,delimiter=',')
header = next(data)

day = [] #일별 데이터를 저장할 리스트 day생성
for i in range(31):
    day.append([]) #dayl리스트 내 31개 리스트 생성
    
for row in data :
    if row[-1] != '':
        if row[0].split('-')[1] == '08': #8월 이라면
            day[int(row[0].split('-')[2])-1].append(float(row[-1])) #최고 기온값 저장
            
plt.style.use('ggplot') #그래프 스타일 지정
plt.figure(figsize=(10,5),dpi=300) #그래프 크기 수정
plt.boxplot(day, showfliers=False) #아웃라이어 값 생략
plt.show()
```
