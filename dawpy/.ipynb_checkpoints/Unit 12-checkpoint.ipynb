```python
import csv
f=open('subwaytime.csv')
data =csv.reader(f)
next(data)
next(data)
mx=0
mx_station = ''
t = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요?:'))

for row in data :
    row[4:-1]=map(int, row[4:-1])
    a= row[4+(t-4)*2]
    if a >mx :
        mx = a
        mx_station = row[3]+'('+row[1]+')'
print(mx_station,mx)
    
```

```python
import csv
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)
mx=[0]*24
mx_station=['']*24
for row in data:
    row[4:-1]=map(int,row[4:-1])
    for j in range(24):
        a = row[j*2+4]
        if a > mx[j]:
            mx[j] = a
            r=j+4
            if r > 24 :  #25,26시 등을 1시, 2로 변경
                r=r-24
            mx_station[j] = row[3]+'('+str(r)+')'
print(mx_station)
print(mx)
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.bar(range(24),mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f= open('subwaytime.csv')
data = csv.reader(f)

next(data)
next(data)

mx=[0]*24
mx_station =['']*24

for row in data :
    row[4:-1]=map(int, row[4:-1])
    for j in range(24):
        b= row[5+j*2]
        if b> mx[j]:
            mx[j]=b
            r=j+4
            if r >24:
                r=r-24
            mx_station[j] = row[3]+'('+str(r)+')'
            
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx, color='b')
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
import csv
f=open('subwaytime.csv')
data =csv.reader(f)
next(data)
next(data)
s_in = [0]*24
s_out = [0]*24

for row in data :
    row[4:-1]=map(int, row[4:-1])
    for i in range(24):
        s_in[i]+=row[4+i*2]
        s_out[i]+=row[5+i*2]
        
plt.rc('font',family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
x=[]
for i in range(4,28) :
    if i > 24:
        i=i-24
    x.append(i)

plt.xticks(range(24),x)
plt.show()
```
