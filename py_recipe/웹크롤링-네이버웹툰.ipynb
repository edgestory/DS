```python
from bs4 import BeautifulSoup
from pprint import pprint
import requests
```

```python
html = requests.get("https://comic.naver.com/webtoon/weekday.nhn")
soup = BeautifulSoup(html.text,'html.parser')
html.close()
```

```python
pprint(soup)
```

```python
data1=soup.find('div',{'class':'col_inner'})
pprint(data1)
```

```python
data2 = data1.findAll('a',{'class':'title'})
pprint(data2)
```

```python
title_list=[]
for t in data2 :
    title_list.append(t.text)
    
pprint(title_list)
```

```python
title_list = []
title_list = [t.text for t in data2]
pprint(title_list)
```

```python
data3_list = soup.findAll('div',{'class':'col_inner'})
pprint(data3_list)
```

```python
week_title_list = []

for data3 in data3_list:
    data4=data3.findAll('a',{'class':'title'})
    
    title_list = [t.text for t in data4]
    #week_title_list.append(title_list)
    week_title_list.extend(title_list)

pprint(week_title_list)
```
