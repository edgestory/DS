### choropleth를 위한 GeoJSON

```python
geo_path = 'seoul_municipalities_geo_simple.json'

import json
geo_json=json.load(open(geo_path, encoding='utf-8'))
```

```python
geo_json
```

```python
geo_json['features'][0]['properties']
```

### 스타벅스 매장 분포

```python
df_star = df_cafe_vs[df_cafe_vs['브랜드명']=='스타벅스'].copy()
df_star.head()
```

```python
geo_json['features'][0]['properties']['name']
```

```python
m=folium.Map([lat,long])

folium.Choropleth(
    geo_data = geo_json,
    name = 'choropleth',
    data=df_star,
    columns=['구','매장수'],
    key_on = 'feature.properties.name',
    fill_color = 'YlGn',
    fill_opacity=0.7,
    line_poacity=0.2,
    legend_name='스타벅스 매장수'

).add_to(m)

m
```

```python
df_vs = df_cafe_vs.pivot(index='구', columns='브랜드명', values='매장수')
```

```python
df_vs.columns
```

```python
df_vs['매장수비교'] =  df_vs['스타벅스'] - df_vs['이디야'] >0
```

```python
df_vs.head()
```

```python
df_vs['매장수비교']=df_vs['매장수비교'].astype(float)
df_vs
```

```python
df_vs = df_vs.reset_index()
df_vs.head()
```

```python
m=folium.Map([lat,long],tiles='Stamen Toner')

folium.Choropleth(
    geo_data = geo_json,
    name = 'choropleth',
    data=df_vs,
    columns=['구','매장수비교'],
    key_on = 'feature.properties.name',
    fill_color = 'BuGn',
    fill_opacity=0.7,
    line_poacity=0.2,
    legend_name='매장수 비교'

).add_to(m)

m
```

```python
for i in df_vs.index:
    print(df_vs.loc[i,'구'])
```

```python
df_gu_mean = df_cafe.pivot_table(index='시군구명'
                                ,values=['위도','경도'], aggfunc='mean')
df_gu_mean = df_gu_mean.reset_index()
df_gu_mean.head()
```

```python
df_vs=df_vs.merge(df_gu_mean, how='left',left_on='구',right_on='시군구명')
```

```python
for i in df_vs.index:
    sub_long = df_vs.loc[i,'경도']
    sub_lat = df_vs.loc[i,'위도']
    
    print(df_vs.loc[i,'구'],sub_long,sub_lat)
```

```python
df_vs.loc[1,'스타벅스']
```

```python
m=folium.Map([lat,long],tiles='Stamen Toner')

folium.Choropleth(
    geo_data = geo_json,
    name = 'choropleth',
    data=df_vs,
    columns=['구','매장수비교'],
    key_on = 'feature.properties.name',
    fill_color = 'BuGn',
    fill_opacity=0.7,
    line_poacity=0.2,
    legend_name='매장수 비교'

).add_to(m)

for i in df_vs.index:
    sub_long = df_vs.loc[i,'경도']
    sub_lat = df_vs.loc[i,'위도']
    
    cafes = ['스타벅스','이디야']
    for cafe in cafes:
        cafe_count = df_vs.loc[i,cafe]
        radius = np.sqrt(cafe_count) * 3
        gu = df_vs.loc[i,'구']
        tootip = f"{gu}{cafe}:{cafe_count}"
        
        color = 'green'
        if cafe=='이디야':
            color='blue'
            sub_long=sub_long + 0.01
        
        folium.CircleMarker([sub_lat,sub_long]
                           ,radius=float(radius)
                           ,color = color
                            ,fill=True
                            ,tooltip = tootip
                           ).add_to(m)

m
m.save('chapter4.html')
```
