### 편익시설

```python
df.head()
```

```python
def get_string(text):
    if pd.isna(text):
        return text
    
    text = re.sub("[\d\.'\-(nan)\[\]]","",text)
    text = re.sub(" ","",text)
    text = re.sub(",{2,}",",",text)
    return text
```

```python
get_string("[1. '화장실' 5, 2. 주차장, - ,nan]")
```

```python
get_string(np.nan)
```

```python
df['편익시설']=df['공원보유시설(편익시설)'].map(get_string)
```

## 특정 정보 마스킹

### 전화번호 마스킹

```python
re.sub("(\d{3})-(\d{2})(\d{2})-(\d{2})(\d{2})",r"\1-\2**-\4**",'010-1234-5678')
```

```python
re.sub("(\d{3})-(\d{1,2})(\d{2})-(\d{2})(\d{2})",r"\1-\2**-\4**",'010-123-5678')
```

```python
re.sub("(\d{3})-(\d{1,2})(\d{2})-(\d{2})(\d{2})",r"\1-\2**-\4**",'+82-10-123-5678')
```

```python
phone = '+82-10-1234-5678'
```

```python
def mask_phone_number(phone):
    if pd.isnull(phone):
        return phone
    
    num=phone.split("-")
    num_len = len(num)//2
    
    num[num_len] = re.sub("\w","*",num[num_len])
    num[-1]=re.sub("(\d{2})(\d{2})",r"**\2",num[-1])
    
    return "-".join(num)

mask_phone_number(phone)
```

```python
mask_phone_number(np.nan)
```

```python
df['전화번호(전처리)']=df['전화번호'].map(mask_phone_number)
```

```python
df[['전화번호(전처리)','전화번호']]
```

```python
def mask_email(email):
    address, domain = email.split("@")
    mask = re.sub("\w",'*', address[1:-1])
    return f"{address[0]}{mask}{address[-1]}@{domain}"
```
